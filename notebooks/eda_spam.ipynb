```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
data = pd.read_csv(r"C:\Users\nagen\Desktop\Kaggle_datasets\SMS Spam collection\Spam-Classification-MLOps-Project\data\smsspamcollection\SMSSpamCollection",
                   sep="\t", header=None, names=["label", "message"])


# Quick look
print("Shape of dataset:", data.shape)
print(data.head())

# Distribution of labels
print("\nLabel counts:\n", data['label'].value_counts())

sns.countplot(x='label', data=data)
plt.title("Spam vs Ham Distribution")
plt.show()

# Example messages
print("\n--- Example Ham Message ---\n", data[data['label']=="ham"]['message'].iloc[0])
print("\n--- Example Spam Message ---\n", data[data['label']=="spam"]['message'].iloc[0])
```
